```python
import pandas as pd
import numpy as np
import pylab
import scipy.stats as scst
import matplotlib.pyplot as plt
import statistics as st
```

```python
data = pd.read_csv('gasoline.csv')
```

```python
scst.probplot(data['y'], dist="norm", plot = pylab)
pylab.show()
```

```python
l = list(np.arange(min(data['y']) - 5, max(data['y']) + 5, 0.5))
ln = scst.norm.pdf(l, loc = st.mean(data['y']), scale = st.stdev(data['y']))
plt.plot(l, ln)
plt.hist(data['y'], density = True)
plt.title("Histogram of Mileage")
plt.show()
```
