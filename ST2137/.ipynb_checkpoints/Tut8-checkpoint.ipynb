```python
import pandas as pd
import numpy as np
import scipy
import math
import statistics as st
import scipy.stats as scst
import statsmodels.api
import matplotlib.pyplot as plt
import pylab
```

```python
flextime = pd.read_csv('./flextime.txt', delimiter = ' ')
```

```python
flextime
```

```python
diff = flextime.after - flextime.before
```

```python
scst.wilcoxon(x = diff, zero_method = 'wilcox', correction = False, alternative = 'greater')
```

```python
gasoline = pd.read_csv('./gasoline.csv')
```

```python
gasoline
```

```python
auto = gasoline[gasoline.x11 == 1].y
```

```python
manual = gasoline[gasoline.x11 == 0].y
```

```python
scst.shapiro(manual)
```

```python
scst.shapiro(auto)
```

```python
scst.bartlett(manual, auto)
```

```python
scst.ttest_ind(manual, auto, axis = 0, equal_var = False, alternative = 'greater')
```
