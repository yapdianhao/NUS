```python
import numpy as np
import pandas as pd
import statistics as st
from statistics import mean
from statistics import median
from statistics import variance
import math

import scipy.stats as scst
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid") 
```

```python
np.random.seed(999)
```

```python
means = np.zeros(100)
```

```python
for i in range(100):
    sample = np.random.exponential(5000, 30)
    means[i] = st.mean(sample)
```

```python
st.mean(means)
```

```python
st.stdev(means)
```

```python
5000 / math.sqrt(30)
```

```python
l = list(np.arange(min(means), max(means), step = (max(means) - min(means)) / 100))
y = scst.norm.pdf(l, loc = st.mean(means), scale = st.stdev(means))
plt.plot(l, y)
plt.hist(means, range = None, density = True, color = 'C4')
plt.title('Histogram of sample means')
plt.show()
```

```python
scst.shapiro(means)
```

```python
np.count_nonzero(means < (st.mean(means) - 2 * st.stdev(means)))
```

```python
np.count_nonzero(means > (st.mean(means) + 2 * st.stdev(means)))
```
