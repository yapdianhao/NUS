```python
machines = pd.read_csv('./machine.txt', delimiter = ' ')
```

```python
machines
```

```python
old = machines[machines['machine'] == 'O']
```

```python
new = machines[machines['machine'] == 'N']
```

```python
old_strength = old.strength
```

```python
new_strength = new.strength
```

```python
scst.bartlett(old_strength, new_strength)
```

```python
scst.ttest_ind(old_strength, new_strength, axis = 0, equal_var = True, alternative = "greater")
```
